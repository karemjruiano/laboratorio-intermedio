```python
import pandas as pd
df_datos_parte_4 = pd.read_csv("Efecto Hall Temperaturas.csv")


columnas=[]
for i in range(0,12):
    columnas.append(str(i))
df_datos_parte_4.columns = columnas

Temp_n=(df_datos_parte_4["2"].iloc[2:]).values.tolist()
Temp_p=(df_datos_parte_4["6"].iloc[2:]).values.tolist()
Temp_g=(df_datos_parte_4["10"].iloc[2:]).values.tolist()
Conductividad_n=(df_datos_parte_4["4"].iloc[2:]).values.tolist()
Conductividad_p=(df_datos_parte_4["7"].iloc[2:]).values.tolist()
Conductividad_g=(df_datos_parte_4["11"].iloc[2:]).values.tolist()
```

```python
import matplotlib.pyplot as plt
```

```python
Temp_n=[float(item) for item in Temp_n]
Temp_p=[float(item) for item in Temp_p]
Temp_g=[float(item) for item in Temp_g]
Conductividad_n=[float(item) for item in Conductividad_n]
Conductividad_p=[float(item) for item in Conductividad_p]
Conductividad_g=[float(item) for item in Conductividad_g]
```

```python
Serie_1=[Temp_n,Conductividad_n]
Serie_2=[Temp_p,Conductividad_p]
Serie_3=[Temp_g,Conductividad_g]
Series=[Serie_1,Serie_2,Serie_3]
```

```python
ax = plt.axes()
labels=["n","p","neutro"]
i=0
for Serie in Series:
    plt.plot(Serie[0], Serie[1],label=labels[i])
    i+=1
    
plt.legend(loc='upper left')
plt.xlabel("Inverso Temperatura (1/°C)")
plt.ylabel("Conductividad")
```
